```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("/content/4_6008063728413576163.csv")
df.head()
```

```python
sizes = df['Parch'].value_counts(sort = 1)
sizes
```

```python
df.info()
```

```python
df.drop(['Cabin'], axis=1, inplace=True)
df.drop(['Name'], axis=1, inplace=True)
df.drop(['Ticket'], axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.Embarked[df.Embarked == 'S'] = 1
df.Embarked[df.Embarked== 'C'] = 2
df.Embarked[df.Embarked=='Q']=3
df.Sex[df.Sex == 'female'] = 1
df.Sex[df.Sex== 'male'] = 0
df.head()
```

```python
Y = df["Parch"]
Y=Y.astype('int')
X = df.drop(labels = ["Embarked"], axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=20)
```

```python
df
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 20 , random_state = 30)
model.fit(X_train, y_train)
```

```python
prediction_test = model.predict(X_test)
print(y_test, prediction_test)
```

```python
from sklearn import metrics
print ("Accuracy = ", metrics.accuracy_score(y_test, prediction_test))
```

```python
feature_list = list(X.columns)
feature_imp = pd.Series(model.feature_importances_,index=feature_list).sort_values(ascending=False)
print(feature_imp)
```
