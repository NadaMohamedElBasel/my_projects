### Import `Numpy` ?<a id='n.1'></a>


```python
import numpy as np
```

### `Create` an array of 10 zeros ?

```python
zero=np.repeat(0,10)
zero
```

### `Create` an array of 10 Ones ?

```python
one=np.repeat(1,10)
one
```

### `Create` an array of 10 Fours ?

```python
four=np.repeat(4,10)
four
```

### `Create` an array of the Integers from 15 to 55 ?

```python
ints=np.arange(15,56,1,int)
ints
```

### `Create` an array of the Even Integers from 15 to 55 ?

```python
evints=np.arange(16,55,2,int)
evints
```

### `Create` a Matrix (3×3) with Values  from 1 to 9 ?

```python
mat1=np.arange(1,10,1,int).reshape(3,3)
mat1
```

### `Create` a random number between 0 and 1 `[0, 1)`?

```python
num1=np.random.random()
num1
```

### `Create` an array of 10 `random` numbers between 0 and 1 `[0, 1)`?

```python
randarr=np.random.random(10)
randarr
```

### `Create` a Matrix of 9 `random` numbers between 0 and 1 `[0, 1)`?

```python
randmat=np.random.random(9).reshape(3,3)
randmat
```

### `Create` a Matrix of 9 `random` numbers?

```python
mat2=np.random.randn(9).reshape(3,3)
mat2
```

### `Create` a Matrix of 9 `random` numbers integers  between `1 and 100`?

```python
mat3=np.random.randint(1,100,9).reshape(3,3)
mat3
```

### `Create` a Array of 12 `random` numbers from a `standard normal distribution`?

```python
std=np.random.normal(0,1,12)
std
```

### `Create` a Matrix of 100 numbers of `sorted`   between `0.01 and 1` ?
