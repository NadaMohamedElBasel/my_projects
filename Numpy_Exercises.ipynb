#### find common values between two arrays

```python
x1 = np.array([0, 2, 2, 2, 4, 5, 1, 4, 8, 5])
x2 = np.array([2, 2, 3, 4, 7, 4, 6, 9, 4, 8])
```

```python
insec=np.intersect1d(x1,x2)
insec
```
