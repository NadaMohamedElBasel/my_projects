```python
x,y,z = np.shape(img)
red = np.zeros((x,y,z),dtype=int)
green = np.zeros((x,y,z),dtype=int)
blue = np.zeros((x,y,z),dtype=int)
for i in range(0,x):
    for j in range(0,y):
        red[i][j][0] = img[i][j][0]
        green[i][j][1]= img[i][j][1]
        blue[i][j][2] = img[i][j][2]
plt.figure(1) 
plt.imshow(red)
plt.figure(2)
plt.imshow(green)
plt.figure(3)
plt.imshow(blue)
```

```python
def rgb_splitter(anyimg):
    rgb_list = ['Reds','Greens','Blues']
    fig, ax = plt.subplots(1, 3, figsize=(10,10))
    for i in range(3):
        ax[i].imshow(anyimg[:,:,i], cmap = rgb_list[i])
rgb_splitter(img)
```

```python
#using average 
def convert2Gray (img2G):

 img2G = cv2.imread('/content/Screenshot_20230130-233517_Google.jpg')
 (row, col) = img2G.shape[0:2]
 for i in range(row):
     for j in range(col):
         img2G[i, j] = sum(img[i, j]) /3
 plt.imshow(img2G)
convert2Gray(img)
```
